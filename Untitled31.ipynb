```python
import numpy as np
import matplotlib.pyplot as plt

# ---------------------------
# User parameters
# ---------------------------
Re = 100.0       # Reynolds number (U*D/nu)
U_in = 1.0       # inlet uniform velocity
rho = 1.0

D = 1.0
Ly = 5.0 * D
Lx = (2.5 + 15.0) * D  # cylinder center 2.5D from inlet, outlet 15D after center -> total ~17.5D
# Slight padding to be safe:
Lx = 18.0 * D

# Grid resolution (adjust if slow)
Nx = 320   # points in x
Ny = 96    # points in y

dx = Lx / (Nx - 1)
dy = Ly / (Ny - 1)

x = np.linspace(0.0, Lx, Nx)
y = np.linspace(0.0, Ly, Ny)
X, Y = np.meshgrid(x, y)

# Cylinder geometry
cx = 2.5 * D
cy = 0.5 * Ly
R  = 0.5 * D

solid = (X - cx)**2 + (Y - cy)**2 <= R**2  # boolean mask for cylinder interior

# Kinematic viscosity from Re
nu = U_in * D / Re

# Time stepping
CFL = 0.25
dt_adv = CFL * min(dx, dy) / max(U_in, 1e-12)
dt_diff = 0.25 * min(dx, dy)**2 / max(nu, 1e-12)
dt = min(dt_adv, dt_diff)

t_end = 40.0  # simulate long enough to see shedding (Re~100)
nsteps = int(t_end / dt)

# Pressure Poisson iterations per time step
pp_iters = 80
omega = 1.7  # SOR relaxation (1.0 = Gauss-Seidel). Try 1.6-1.9.

# Output control
plot_every = max(1, nsteps // 80)

print(f"Grid: {Nx} x {Ny}, dx={dx:.4f}, dy={dy:.4f}")
print(f"Re={Re}, nu={nu:.6f}, dt={dt:.6f}, steps={nsteps}")

# ---------------------------
# Helper finite-difference ops
# ---------------------------

def ddx_central(f, dx):
    """Central difference in x for interior points."""
    return (f[:, 2:] - f[:, :-2]) / (2.0 * dx)

def ddy_central(f, dy):
    """Central difference in y for interior points."""
    return (f[2:, :] - f[:-2, :]) / (2.0 * dy)

def laplacian(f, dx, dy):
    """Laplacian for interior points."""
    return ((f[1:-1, 2:] - 2*f[1:-1, 1:-1] + f[1:-1, :-2]) / dx**2 +
            (f[2:, 1:-1] - 2*f[1:-1, 1:-1] + f[:-2, 1:-1]) / dy**2)

def apply_bc(u, v, p):
    """
    Boundary conditions:
      Inlet (x=0): u=U_in, v=0
      Outlet (x=Lx): du/dx=0, dv/dx=0, p=0 (reference)
      Walls (y=0, y=Ly): no-slip u=v=0, dp/dn=0
      Cylinder (mask): u=v=0
    """
    # Inlet
    u[:, 0] = U_in
    v[:, 0] = 0.0

    # Outlet: zero-gradient for u,v, and p=0
    u[:, -1] = u[:, -2]
    v[:, -1] = v[:, -2]
    p[:, -1] = 0.0

    # Walls: no-slip
    u[0, :] = 0.0
    v[0, :] = 0.0
    u[-1, :] = 0.0
    v[-1, :] = 0.0

    # Pressure Neumann at walls: dp/dy=0 -> copy neighbor
    p[0, :] = p[1, :]
    p[-1, :] = p[-2, :]

    # Cylinder no-slip
    u[solid] = 0.0
    v[solid] = 0.0

    return u, v, p

def build_rhs_pressure(u_star, v_star, dx, dy, dt):
    """
    RHS for pressure Poisson:
      p = (rho/dt) * div(u*)
    We'll solve for p with rho=1 in nondimensional form.
    """
    div = ((u_star[1:-1, 2:] - u_star[1:-1, :-2])/(2*dx) +
           (v_star[2:, 1:-1] - v_star[:-2, 1:-1])/(2*dy))
    rhs = (rho/dt) * div
    return rhs

def pressure_poisson_sor(p, rhs, dx, dy, iters=80, omega=1.7):
    """
    Solve p = rhs on interior using SOR.
    Boundary conditions handled outside this function by apply_bc.
    """
    dx2 = dx*dx
    dy2 = dy*dy
    denom = 2.0*(dx2 + dy2)

    for _ in range(iters):
        # Gauss-Seidel + SOR update on interior
        p_old = p.copy()

        p[1:-1, 1:-1] = (1-omega)*p[1:-1, 1:-1] + omega * (
            ( (p[1:-1, 2:] + p[1:-1, :-2]) * dy2 +
              (p[2:, 1:-1] + p[:-2, 1:-1]) * dx2 -
              rhs * dx2 * dy2
            ) / denom
        )

        # Optional: keep pressure inside cylinder reasonable
        # (copy from nearest fluid neighbor using a simple "fill")
        p[solid] = 0.0  # gauge-ish; not physically perfect but stabilizes

    return p

def compute_vorticity(u, v, dx, dy):
    """ = dv/dx - du/dy (interior). Return full field with zeros at boundaries."""
    w = np.zeros_like(u)
    dvdx = (v[:, 2:] - v[:, :-2])/(2*dx)
    dudy = (u[2:, :] - u[:-2, :])/(2*dy)
    w[1:-1, 1:-1] = dvdx[1:-1, :] - dudy[:, 1:-1]
    w[solid] = 0.0
    return w

# ---------------------------
# Initialization
# ---------------------------
u = np.zeros((Ny, Nx))
v = np.zeros((Ny, Nx))
p = np.zeros((Ny, Nx))

# Start with inlet field everywhere to speed up transient
u[:, :] = U_in
v[:, :] = 0.0
u[0, :] = 0.0
u[-1, :] = 0.0
u[solid] = 0.0

u, v, p = apply_bc(u, v, p)

# For force estimate (penalization-like)
drag_hist = []
lift_hist = []
time_hist = []

cell_area = dx * dy
solid_area = solid.sum() * cell_area

# ---------------------------
# Main time loop
# ---------------------------
for n in range(nsteps):
    t = (n+1)*dt

    # Save previous
    un = u.copy()
    vn = v.copy()

    # ---------- Advection (use a simple first-order upwind for robustness) ----------
    # Upwind derivatives for u and v:
    dudx = np.zeros_like(u)
    dudy = np.zeros_like(u)
    dvdx = np.zeros_like(v)
    dvdy = np.zeros_like(v)

    # x-derivatives (upwind)
    dudx[:, 1:-1] = np.where(un[:, 1:-1] > 0,
                            (un[:, 1:-1] - un[:, :-2]) / dx,
                            (un[:, 2:] - un[:, 1:-1]) / dx)
    dvdx[:, 1:-1] = np.where(un[:, 1:-1] > 0,
                            (vn[:, 1:-1] - vn[:, :-2]) / dx,
                            (vn[:, 2:] - vn[:, 1:-1]) / dx)

    # y-derivatives (upwind)
    dudy[1:-1, :] = np.where(vn[1:-1, :] > 0,
                             (un[1:-1, :] - un[:-2, :]) / dy,
                             (un[2:, :] - un[1:-1, :]) / dy)
    dvdy[1:-1, :] = np.where(vn[1:-1, :] > 0,
                             (vn[1:-1, :] - vn[:-2, :]) / dy,
                             (vn[2:, :] - vn[1:-1, :]) / dy)

    adv_u = un * dudx + vn * dudy
    adv_v = un * dvdx + vn * dvdy

    # ---------- Diffusion (Laplacian, central) ----------
    lap_u = np.zeros_like(u)
    lap_v = np.zeros_like(v)
    lap_u[1:-1, 1:-1] = laplacian(un, dx, dy)
    lap_v[1:-1, 1:-1] = laplacian(vn, dx, dy)

    # ---------- Intermediate velocity u* (no pressure) ----------
    u_star = un + dt * (-adv_u + nu * lap_u)
    v_star = vn + dt * (-adv_v + nu * lap_v)

    # Enforce BC on intermediate
    u_star, v_star, p = apply_bc(u_star, v_star, p)

    # Enforce cylinder no-slip strongly (mask)
    u_star[solid] = 0.0
    v_star[solid] = 0.0

    # ---------- Pressure Poisson ----------
    rhs = build_rhs_pressure(u_star, v_star, dx, dy, dt)
    p = pressure_poisson_sor(p, rhs, dx, dy, iters=pp_iters, omega=omega)
    u_star, v_star, p = apply_bc(u_star, v_star, p)

    # ---------- Projection step ----------
    dpdx = np.zeros_like(p)
    dpdy = np.zeros_like(p)
    dpdx[:, 1:-1] = (p[:, 2:] - p[:, :-2])/(2*dx)
    dpdy[1:-1, :] = (p[2:, :] - p[:-2, :])/(2*dy)

    u = u_star - (dt/rho) * dpdx
    v = v_star - (dt/rho) * dpdy

    # Final BC
    u, v, p = apply_bc(u, v, p)

    # ---------- Approx drag/lift (mask-based forcing) ----------
    # Since we "force" velocities in solid to 0, estimate the force as:
    #   F    (u_star - u_new)/dt dV  (momentum removed per time)
    # Here u_new is u after projection & BC.
    # Use the solid region as the "penalized" zone.
    Fx = rho * np.sum((u_star[solid] - u[solid]) / dt) * cell_area
    Fy = rho * np.sum((v_star[solid] - v[solid]) / dt) * cell_area

    # Convert to coefficients with reference: 0.5*rho*U^2*D (2D force per unit depth)
    # In 2D, area reference is D*1 (unit depth), so:
    q = 0.5 * rho * U_in**2
    Cd = Fx / (q * D)
    Cl = Fy / (q * D)

    drag_hist.append(Cd)
    lift_hist.append(Cl)
    time_hist.append(t)

    # ---------- Plot ----------
    if (n % plot_every == 0) or (n == nsteps-1):
        w = compute_vorticity(u, v, dx, dy)

        plt.figure(figsize=(12, 4))
        plt.subplot(1, 3, 1)
        plt.title(f"u-velocity (t={t:.2f})")
        plt.imshow(u, origin="lower", extent=[0, Lx, 0, Ly], aspect="auto")
        plt.colorbar()

        plt.subplot(1, 3, 2)
        plt.title("vorticity ")
        plt.imshow(w, origin="lower", extent=[0, Lx, 0, Ly], aspect="auto")
        plt.colorbar()

        plt.subplot(1, 3, 3)
        plt.title("Cylinder + streamlines (coarse)")
        # streamplot needs structured grid with increasing x,y
        skip = (slice(None, None, 3), slice(None, None, 3))
        plt.streamplot(X[skip], Y[skip], u[skip], v[skip], density=1.2, linewidth=0.7, arrowsize=0.7)
        # draw cylinder
        th = np.linspace(0, 2*np.pi, 200)
        plt.plot(cx + R*np.cos(th), cy + R*np.sin(th))
        plt.xlim(0, Lx)
        plt.ylim(0, Ly)

        plt.tight_layout()
        plt.show()

# ---------------------------
# Post-processing: drag/lift time series
# ---------------------------
plt.figure(figsize=(10,4))
plt.plot(time_hist, drag_hist, label="Cd (approx)")
plt.plot(time_hist, lift_hist, label="Cl (approx)")
plt.xlabel("time")
plt.ylabel("coeff")
plt.grid(True)
plt.legend()
plt.show()

print("Done.")
print(f"Cd mean (last 20%): {np.mean(drag_hist[int(0.8*len(drag_hist)):]):.3f}")
print(f"Cl rms  (last 20%): {np.sqrt(np.mean(np.array(lift_hist[int(0.8*len(lift_hist)):])**2)):.3f}")
```
