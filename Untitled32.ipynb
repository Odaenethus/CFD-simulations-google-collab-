```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation

# =========================
# 1) Domain & parameters
# =========================
Re  = 100.0
Uin = 1.0
rho = 1.0
D   = 1.0
R   = 0.5*D

Ly = 5.0*D
Lx = 18.0*D

# Higher resolution for better visuals
Nx = 520   # try 420 if slow
Ny = 180   # try 140 if slow

dx = Lx/(Nx-1)
dy = Ly/(Ny-1)

x = np.linspace(0, Lx, Nx)
y = np.linspace(0, Ly, Ny)
X, Y = np.meshgrid(x, y)

cx, cy = 2.5*D, 0.5*Ly
r = np.sqrt((X-cx)**2 + (Y-cy)**2)

# Brinkman penalization: smooth mask near cylinder boundary
# chi ~ 1 inside cylinder, 0 outside with a thin transition
transition = 2.0*max(dx,dy)
chi = 0.5*(1.0 - np.tanh((r - R)/transition))  # smooth step

nu = Uin*D/Re

# Time step
CFL = 0.35
dt_adv  = CFL*min(dx,dy)/Uin
dt_diff = 0.25*min(dx,dy)**2/nu
dt = min(dt_adv, dt_diff)

t_end = 120.0  # longer -> clearer vortex street
nsteps = int(t_end/dt)

print(f"Nx={Nx}, Ny={Ny}, dx={dx:.4f}, dy={dy:.4f}, dt={dt:.6f}, steps={nsteps}, nu={nu:.6f}")

# Pressure solver params
pp_iters = 180
omega_sor = 1.7

# Brinkman penalization strength (large -> enforce no-slip strongly)
eta = 1e-3  # smaller => stronger forcing (but too small can destabilize)

# Mild stabilization for central advection (like "numerical viscosity")
eps_art = 0.01

# =========================
# 2) Operators
# =========================
def ddx(f):
    g = np.zeros_like(f)
    g[:, 1:-1] = (f[:, 2:] - f[:, :-2])/(2*dx)
    return g

def ddy(f):
    g = np.zeros_like(f)
    g[1:-1, :] = (f[2:, :] - f[:-2, :])/(2*dy)
    return g

def lap(f):
    g = np.zeros_like(f)
    g[1:-1, 1:-1] = (
        (f[1:-1, 2:] - 2*f[1:-1, 1:-1] + f[1:-1, :-2])/dx**2 +
        (f[2:, 1:-1] - 2*f[1:-1, 1:-1] + f[:-2, 1:-1])/dy**2
    )
    return g

def apply_bc(u, v, p):
    # Inlet: uniform + small perturbation to trigger shedding
    u[:, 0] = Uin
    v[:, 0] = 2e-3*np.sin(2*np.pi*y/Ly)  # tiny symmetry breaker

    # Outlet: zero-gradient for u,v; set p=0 reference
    u[:, -1] = u[:, -2]
    v[:, -1] = v[:, -2]
    p[:, -1] = 0.0

    # Walls (no-slip)
    u[0, :], v[0, :] = 0.0, 0.0
    u[-1, :], v[-1, :] = 0.0, 0.0

    # Pressure Neumann at walls + copy at inlet
    p[0, :]  = p[1, :]
    p[-1, :] = p[-2, :]
    p[:, 0]  = p[:, 1]
    return u, v, p

def build_rhs(u_star, v_star):
    div = ddx(u_star) + ddy(v_star)
    return (rho/dt)*div[1:-1, 1:-1]

def pressure_poisson_sor(p, rhs):
    dx2, dy2 = dx*dx, dy*dy
    denom = 2*(dx2 + dy2)

    for _ in range(pp_iters):
        p[1:-1, 1:-1] = (1-omega_sor)*p[1:-1, 1:-1] + omega_sor * (
            ((p[1:-1, 2:] + p[1:-1, :-2])*dy2 +
             (p[2:, 1:-1] + p[:-2, 1:-1])*dx2 -
             rhs*dx2*dy2) / denom
        )

        # enforce pressure BC quickly each sweep
        p[:, -1] = 0.0
        p[:, 0]  = p[:, 1]
        p[0, :]  = p[1, :]
        p[-1, :] = p[-2, :]

    return p

def vorticity(u, v):
    w = ddx(v) - ddy(u)
    return w

# =========================
# 3) Initialize
# =========================
u = np.ones((Ny, Nx))*Uin
v = np.zeros((Ny, Nx))
p = np.zeros((Ny, Nx))
u, v, p = apply_bc(u, v, p)

# =========================
# 4) Visualization helpers
# =========================
def draw_snapshot(u, v, t, levels=60):
    w = vorticity(u, v)

    fig = plt.figure(figsize=(14, 4))

    # Vorticity (contours look "textbook")
    ax1 = plt.subplot(1, 2, 1)
    ax1.set_title(f"Vorticity  (t={t:.2f})")
    # focus range based on percentiles for clearer contrast
    w_clip = np.percentile(np.abs(w), 98)
    cf = ax1.contourf(X, Y, w, levels=levels, vmin=-w_clip, vmax=w_clip)
    plt.colorbar(cf, ax=ax1)
    # cylinder
    th = np.linspace(0, 2*np.pi, 400)
    ax1.fill(cx + R*np.cos(th), cy + R*np.sin(th), color="0.6")
    ax1.set_xlim(0, Lx); ax1.set_ylim(0, Ly)

    # Streamlines
    ax2 = plt.subplot(1, 2, 2)
    ax2.set_title("Streamlines")
    skip = (slice(None, None, 2), slice(None, None, 2))
    ax2.streamplot(X[skip], Y[skip], u[skip], v[skip],
                   density=2.2, linewidth=1.0, arrowsize=1.0)
    ax2.fill(cx + R*np.cos(th), cy + R*np.sin(th), color="0.6")
    ax2.set_xlim(0, Lx); ax2.set_ylim(0, Ly)

    plt.tight_layout()
    plt.show()

# =========================
# 5) Time integration (Projection + Brinkman penalization)
# =========================
# Save frames for animation (optional)
save_animation = True
frame_every = max(1, nsteps//300)  # ~300 frames max
frames = []

for n in range(nsteps):
    t = (n+1)*dt

    un, vn = u.copy(), v.copy()

    # Advection (central)
    adv_u = un*ddx(un) + vn*ddy(un)
    adv_v = un*ddx(vn) + vn*ddy(vn)

    # Diffusion (+ small stabilization)
    diff_u = (nu + eps_art)*lap(un)
    diff_v = (nu + eps_art)*lap(vn)

    # Brinkman penalization forcing toward solid velocity (0,0)
    # du/dt += -(chi/eta) * u
    force_u = -(chi/eta)*un
    force_v = -(chi/eta)*vn

    # Intermediate velocity
    u_star = un + dt*(-adv_u + diff_u + force_u)
    v_star = vn + dt*(-adv_v + diff_v + force_v)

    u_star, v_star, p = apply_bc(u_star, v_star, p)

    # Pressure Poisson
    rhs = build_rhs(u_star, v_star)
    p = pressure_poisson_sor(p, rhs)

    # Projection
    dpdx = ddx(p)
    dpdy = ddy(p)
    u = u_star - (dt/rho)*dpdx
    v = v_star - (dt/rho)*dpdy

    u, v, p = apply_bc(u, v, p)

    # Store frames
    if save_animation and (n % frame_every == 0):
        frames.append((u.copy(), v.copy(), t))

    # A few good snapshots while running
    if n in [int(30/dt), int(60/dt), int(90/dt), int(120/dt)-1]:
        draw_snapshot(u, v, t)

print("Simulation done. Frames:", len(frames))

# =========================
# 6) Optional: animation (streamlines + vorticity)
# =========================
if save_animation and len(frames) > 5:
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 4))

    th = np.linspace(0, 2*np.pi, 400)
    cylx = cx + R*np.cos(th)
    cyly = cy + R*np.sin(th)

    # initialize
    u0, v0, t0 = frames[0]
    w0 = vorticity(u0, v0)
    w_clip0 = np.percentile(np.abs(w0), 98)

    cf = ax1.contourf(X, Y, w0, levels=60, vmin=-w_clip0, vmax=w_clip0)
    cb = fig.colorbar(cf, ax=ax1)
    ax1.fill(cylx, cyly, color="0.6")
    ax1.set_xlim(0, Lx); ax1.set_ylim(0, Ly)
    ax1.set_title(f"Vorticity  (t={t0:.2f})")

    ax2.fill(cylx, cyly, color="0.6")
    ax2.set_xlim(0, Lx); ax2.set_ylim(0, Ly)
    ax2.set_title("Streamlines")

    def update(k):
        ax1.clear(); ax2.clear()
        uk, vk, tk = frames[k]
        wk = vorticity(uk, vk)
        w_clip = np.percentile(np.abs(wk), 98)

        ax1.contourf(X, Y, wk, levels=60, vmin=-w_clip, vmax=w_clip)
        ax1.fill(cylx, cyly, color="0.6")
        ax1.set_xlim(0, Lx); ax1.set_ylim(0, Ly)
        ax1.set_title(f"Vorticity  (t={tk:.2f})")

        skip = (slice(None, None, 2), slice(None, None, 2))
        ax2.streamplot(X[skip], Y[skip], uk[skip], vk[skip],
                       density=2.2, linewidth=1.0, arrowsize=1.0)
        ax2.fill(cylx, cyly, color="0.6")
        ax2.set_xlim(0, Lx); ax2.set_ylim(0, Ly)
        ax2.set_title("Streamlines")

        return []

    anim = FuncAnimation(fig, update, frames=len(frames), interval=50, blit=False)
    plt.tight_layout()
    plt.show()
```
